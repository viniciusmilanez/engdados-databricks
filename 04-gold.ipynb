```python
# =========================================================
# 04_gold
# MODELAGEM DIMENSIONAL
# =========================================================

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.gold")


# =========================================================
# DIM_CLIENTE
# =========================================================

df_clientes = spark.read.table("workspace.silver.clientes")

dim_cliente = df_clientes.select(
    "id_cliente",
    "nome",
    "cidade",
    "estado"
).dropDuplicates()

dim_cliente.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.dim_cliente")


# =========================================================
# DIM_PRODUTO
# =========================================================

df_produtos = spark.read.table("workspace.silver.produtos")

dim_produto = df_produtos.select(
    "id_produto",
    "nome",
    "categoria",
    "preco"
).dropDuplicates()

dim_produto.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.dim_produto")


# =========================================================
# FATO_VENDAS
# =========================================================

df_vendas = spark.read.table("workspace.silver.vendas")

fato_vendas = df_vendas.select(
    "id_venda",
    "id_cliente",
    "id_produto",
    "quantidade",
    "valor_unitario",
    "data_venda"
)

fato_vendas.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.fato_vendas")

print("✅ Camada GOLD criada")
```
