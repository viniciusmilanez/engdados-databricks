```python
# =========================================================
# 01_landing
# EXTRAÇÃO → LANDING/DADOS
# =========================================================

# ---------------------------------------------------------
# 1. Criar schema LANDING
# ---------------------------------------------------------

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.landing")

# ---------------------------------------------------------
# 2. Criar volume
# ---------------------------------------------------------

spark.sql("""
CREATE VOLUME IF NOT EXISTS workspace.landing.dados
""")

# =========================================================
# 3. CLIENTES
# =========================================================

df_clientes = spark.createDataFrame([
    (1, "Ana Silva", "Florianópolis", "SC"),
    (2, "Bruno Costa", "São Paulo", "SP"),
    (3, "Carla Mendes", "Curitiba", "PR"),
    (4, "Diego Lima", "Porto Alegre", "RS"),
    (5, "Eva Souza", "Belo Horizonte", "MG")
], ["id_cliente", "nome", "cidade", "estado"])

# Salvar CSV
df_clientes.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/workspace/landing/dados/clientes")


# =========================================================
# 4. PRODUTOS
# =========================================================

df_produtos = spark.createDataFrame([
    (1, "Notebook Dell", "Informática", 3500.00),
    (2, "Mouse Logitech", "Informática", 120.00),
    (3, "Teclado Mecânico", "Informática", 800.00),
    (4, "Monitor LG 24", "Informática", 1500.00),
    (5, "iPhone 15", "Celular", 5500.00)
], ["id_produto", "nome", "categoria", "preco"])

# Salvar CSV
df_produtos.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/workspace/landing/dados/produtos")


# =========================================================
# 5. VENDAS
# =========================================================

df_vendas = spark.createDataFrame([
    (1, 1, 1, 1, 3500.00, "2024-01-15"),
    (2, 2, 2, 3, 120.00, "2024-01-16"),
    (3, 3, 3, 1, 800.00, "2024-01-17"),
    (4, 1, 4, 2, 1500.00, "2024-01-18"),
    (5, 4, 5, 1, 5500.00, "2024-01-19")
], ["id_venda", "id_cliente", "id_produto",
    "quantidade", "valor_unitario", "data_venda"])

# Salvar CSV
df_vendas.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/workspace/landing/dados/vendas")


print("✅ Arquivos CSV gravados na LANDING")
```
