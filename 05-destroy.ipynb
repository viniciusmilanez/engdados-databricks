```python
# =========================================================
# 99_destroy_environment
# Remove todos os schemas e tabelas do projeto
# =========================================================

print("🧹 Iniciando limpeza do ambiente...")


# =========================================================
# 1. Remover tabelas GOLD
# =========================================================

spark.sql("DROP TABLE IF EXISTS gold.fato_vendas")
spark.sql("DROP TABLE IF EXISTS gold.dim_cliente")
spark.sql("DROP TABLE IF EXISTS gold.dim_produto")

print("✅ Tabelas GOLD removidas")


# =========================================================
# 2. Remover tabelas SILVER
# =========================================================

spark.sql("DROP TABLE IF EXISTS silver.vendas")
spark.sql("DROP TABLE IF EXISTS silver.produtos")
spark.sql("DROP TABLE IF EXISTS silver.clientes")

print("✅ Tabelas SILVER removidas")


# =========================================================
# 3. Remover tabelas BRONZE
# =========================================================

spark.sql("DROP TABLE IF EXISTS bronze.vendas")
spark.sql("DROP TABLE IF EXISTS bronze.produtos")
spark.sql("DROP TABLE IF EXISTS bronze.clientes")

print("✅ Tabelas BRONZE removidas")


# =========================================================
# 4. Remover tabelas LANDING
# =========================================================

spark.sql("DROP TABLE IF EXISTS landing.vendas")
spark.sql("DROP TABLE IF EXISTS landing.produtos")
spark.sql("DROP TABLE IF EXISTS landing.clientes")

print("✅ Tabelas LANDING removidas")


# =========================================================
# 5. Remover schemas
# =========================================================

spark.sql("DROP SCHEMA IF EXISTS gold CASCADE")
spark.sql("DROP SCHEMA IF EXISTS silver CASCADE")
spark.sql("DROP SCHEMA IF EXISTS bronze CASCADE")
spark.sql("DROP SCHEMA IF EXISTS landing CASCADE")

print("✅ Schemas removidos")


# =========================================================
# 6. Validação final
# =========================================================

print("\n📋 Schemas restantes:")
spark.sql("SHOW SCHEMAS").show(truncate=False)

print("✅ Ambiente destruído com sucesso!")
```
