```python
# =========================================================
# 02_bronze
# INGESTÃO → CSV para DELTA
# =========================================================

# ---------------------------------------------------------
# 1. Criar schema BRONZE
# ---------------------------------------------------------

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze")


# =========================================================
# 2. CLIENTES
# =========================================================

df_clientes = spark.read.csv(
    "/Volumes/workspace/landing/dados/clientes",
    header=True,
    inferSchema=True
)

df_clientes.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.bronze.clientes")


# =========================================================
# 3. PRODUTOS
# =========================================================

df_produtos = spark.read.csv(
    "/Volumes/workspace/landing/dados/produtos",
    header=True,
    inferSchema=True
)

df_produtos.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.bronze.produtos")


# =========================================================
# 4. VENDAS
# =========================================================

df_vendas = spark.read.csv(
    "/Volumes/workspace/landing/dados/vendas",
    header=True,
    inferSchema=True
)

df_vendas.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.bronze.vendas")


print("✅ Tabelas BRONZE criadas")
```
